```python
import matplotlib as plt
```

```python
import json
CAPTION_PATH = "../data/image_retrieval/generated_captions/augmented_0.0.json"
THRESHOLD = 0.8
def load_captions(caption_path, threshold=0.8):
    with open(caption_path, 'r') as f:
        caption_list = json.load(f)
    filtered = []
    for elem in caption_list:
        if elem["similarity"] >= threshold:
            filtered.append(elem)
    return filtered

selected_captions = load_captions(CAPTION_PATH, THRESHOLD)
```

```python
%matplotlib inline
import os
BASE_PATH = "/home/deokhk/coursework/"

from matplotlib import pyplot as plt
import matplotlib.image as mpimg


def visualize(caption, base_path):
    candidate_path = os.path.join(base_path, caption["candidate"])
    target_path = os.path.join(base_path, caption["target"])
    similarity = caption["similarity"]
    captions = caption["captions"]
    candidate_img = mpimg.imread(candidate_path)
    target_img = mpimg.imread(target_path)

    f, axarr = plt.subplots(1,2)
    axarr[0].imshow(candidate_img)
    axarr[0].set_title("candidate")
    axarr[1].imshow(target_img)
    axarr[1].set_title("target")
    caption_str = " ".join(captions)
    plt.text(-100, 350, caption_str)
    plt.text(150, 60, str(similarity))

```

```python
for i, elem in enumerate(selected_captions):
    visualize(elem, BASE_PATH)
    if i == 5:
        break
```
